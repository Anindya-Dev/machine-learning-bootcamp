```python
## try, except , else block
try:
    num=int(input("Enter a number: "))
    result=10/num
except ValueError:
    print("That's not a valid number!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
```

```python
## try, except ,else and finally
try:
    num=int(input("Enter a number: "))
    result=10/num
except ValueError:
    print("That's not a valid number!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
finally:
    print("Excution complete")
```
