```python
## Sets

## Sets are a built-in data type in Python used to store collections of unique items. They are unordered , meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection , difference , and symmetric difference.
```

```python
## Create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_empty_set=set([1,2,3,4,5,6,7])
print((my_empty_set))
```

```python
my_empty_set=set([1,2,2,3,4,5,6,7])
print((my_empty_set))
```

```python
## Basics Sets Operation
## Adding and removing Elements

my_set.add(8)
print(my_set)
```

```python
my_set.remove(2)
print(my_set)
```

```python
my_set.remove(11)
print(my_set)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## pop method
removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
## Set Membership test
my_set=(1,2,3,4,5)
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Union
union_set=set1.union(set2)
print(union_set)

## Intersection

intersection_set=set1.intersection(set2)
print(intersection_set)

print(set1)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference
print(set2.difference(set1))
set2.difference_update(set1)
print(set2)
```

```python
## Symmetric Difference
## From set 1 and set 2 which are the common elements only that will be removed and the unique elements will be combiled, basically the intersection elements will be removed

set1.symmetric_difference(set2)
```

```python
## Set Methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```
