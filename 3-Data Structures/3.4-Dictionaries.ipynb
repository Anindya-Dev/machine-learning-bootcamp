```python
## Dictionaries

# Dictionaries are unordered collections of items. They store data in key value pairs. Keys must be unique and immutable(e.g.,strings,numbers,or tuples), while values can be of any type.
```

```python
## Creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"Anindya","age":22,"grade":24}
print(student)
print(type(student))
```

```python
# Keys should be unique
# student={"name":"Anindya","age":32,"name":24}
# print(student)
```

```python
# Accessing Dictionary Elements
print(student['grade'])
```

```python
print(student['age'])
```

```python
# Accessing using get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
# Modifying DIctionary Elements
# Dictionary are mutable, so you can add, update or delete elements
print(student)
```

```python
student['age']=24
print(student)
student["address"]="India"
print(student)
```

```python
del student['grade'] ##d delete key and value pair
print(student)
```

```python
# Dictionary methods

keys=student.keys() ## get all the keys
print(keys)
values=student.values() ## get all the values
print(values)

items=student.items() ## get all key value pairs
print(items)
```

```python
## shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Anindya1"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="Anindya3"
print(student_copy1)
print(student)
```

```python
## Iterating over Dictionaries
## You can use loops to iterate over dictionaries ,keys, values or items(combination of key value pairs)

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values 
```

```python
for value in student.values():
    print(value)
```

```python
for key,value in student.items():
    print(f'{key}:{value}')
```

```python
## Nested Dictionaries 

students={
    "student1":{"name":"Anindya","age":32},
    "student2":{"name":"Peter","age":35}
}
print(students)
```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
```

```python
## Iterating over nested dictionaries

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension
squares={x:x**2 for x in range (5)}
print(squares)
```

```python
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples
## Use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4]
frequency={}
for number in numbers:
    if number in frequency :
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
