```python
## Default Parameters

def greet(name):
    print(f"Welcome,{name}")

greet("Anindya")
```

```python
def greet(name="Guest"):
    print(f"Welcome,{name}")

greet()
```

```python
# Variable length arguments
# Positional and Keyword arguments

def print_numbers(*Anindya):
    for number in Anindya:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Anindya")
```

```python
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"args")
```

```python
## Keyword Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Anindya",age="22",country="India")
```

```python

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional arguments:{val}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Anindya", name="Anindya",age="22",country="India")
```
